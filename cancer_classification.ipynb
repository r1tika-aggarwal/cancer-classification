```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_recall_fscore_support, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("Cancer_Data - Cancer_Data.csv")
```

```python
df.sample(10)
```

```python
df.isna().sum().sort_values()
```

```python
df = df.drop("id", axis = 1) 
```

```python
s = (df.dtypes == 'object')
objcol = list(s[s].index)

print("categorical values-", objcol)
```

```python
from sklearn.preprocessing import OrdinalEncoder
dfnew = df.copy()
ordinal_encoder = OrdinalEncoder()
dfnew[objcol] = ordinal_encoder.fit_transform(df[objcol])
```

```python
dfnew.head()
```

```python
features = dfnew.corr()['diagnosis'].sort_values()
features
```

```python
high_features = features[features >= 0.6].index
high_features
```

```python
X = dfnew[['concavity_worst', 'concavity_mean', 'area_mean', 'radius_mean',
       'area_worst', 'perimeter_mean', 'radius_worst', 'concave points_mean',
       'perimeter_worst', 'concave points_worst']]
y = dfnew['diagnosis']
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
print(f"The range of feature inputs are within {X.min()} to {X.max()}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=125)
```

```python
model = SVC(C=2)
model.fit(X_train, y_train.values.ravel())
preds = model.predict(X_test)
```

```python
 def evaluate_metrics(yt, yp):
    results_pos = {}
    results_pos['accuracy'] = accuracy_score(yt, yp)
    precision, recall, f_beta, _ = precision_recall_fscore_support(yt, yp, average='binary')
    results_pos['recall'] = recall
    results_pos['precision'] = precision
    results_pos['f1score'] = f_beta
    return results_pos

evaluate_metrics(y_test, preds)
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [2, 3, 4, 5.6],
    'gamma': ['scale', 'auto', 0.1, 1.0],
    'coef0': [0.0, 0.1, 0.5, 1.0,2.0,2.3],
    'shrinking': [True, False],
    'tol': [1e-3, 1e-4, 1e-5],
    'max_iter': [-1, 1000, 5000]
}
```

```python
svm_grid_search = GridSearchCV(SVC(), param_grid, cv=5)
svm_grid_search.fit(X_train, y_train)
```

```python
print("\nSVM - Best Parameters:", svm_grid_search.best_params_)
print("SVM - Best Score:", svm_grid_search.best_score_)
```

```python
svm_best_model = svm_grid_search.best_estimator_ #best model
y_pred_svm = svm_best_model.predict(X_test)
print("SVM - Test Accuracy:", accuracy_score(y_test, y_pred_svm))
print(f"SVM: {accuracy_score(y_test, y_pred_svm)}")
```
